```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.read_csv('share.csv')
```

```python
file.head()
```

```python
file=file.replace('*',np.nan)
file
```

```python
file = file.replace('**',np.nan)
file
```

```python
file=file.dropna()
file.tail()
```

```python
#Now getting predictor/feature variables and target variable
X = file['Age-Adjusted Death Rate'].values
X
```

```python
Y = file['Lower 95% Confidence Interval for Death Rate'].values
Y
```

```python
X=X.astype('float')
Y = Y.astype('float')
X_mean = np.mean(X)
Y_mean = np.mean(Y)
n = len(X)
print(n)
```

```python
print(X_mean)
print(Y_mean)
```

```python
num=0
dem=0
for i in range(n):
    num += (X[i]-X_mean)*(Y[i]-Y_mean)
    dem += (X[i]-X_mean)**2
b1 = num/dem
b0 = Y_mean - (X_mean * b1)
print(b0)
print(b1)
```

```python
sns.scatterplot(X,Y,color='red')
x  = np.linspace(min(X)-10,max(X)+10,1000)
y = b1 * x + b0
sns.lineplot(x,y)
plt.xlabel('Age-Adjusted Death Rate')
plt.ylabel('Lower 95% Confidence Interval for Death Rate')
plt.show()
```

```python
rmse=0
for i in range(n):
    y_pred = b0 + b1 * X[i]
    rmse+=(Y[i]-y_pred)**2
rmse=np.sqrt(rmse/n)
rmse
```
